```python
# Project Scope:
# Using data collected by the World Health Organization, I will examine the life expectancy and GDP
#   of a set of countries (Chile, China, Germany, Mexico, US, and Zimbabwe) from 2000 to 2015
# After getting a sense of what the data looks like, I will plot the change in life expectancy
#   for each country over the time period
# I will do the same for their respective GDPs
# I will also plot the averages per country for these metrics alongside one another.
# Finally, I will plot life expectancy and GDP on the x and y axes of a scatter plot incorder
#   to viually get a sense of any correlation
# Along with this I will calculate the pearson correlation for the two variables and its p-value
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.stats import pearsonr
```

```python
df = pd.read_csv("all_data.csv")
df.rename(columns={"Life expectancy at birth (years)": "Life Expectancy"}, inplace=True)
print(df.head())
print(df["Year"].unique())
print(df["Country"].unique())
```

```python
chile_df = df[df["Country"] == "Chile"]
china_df = df[df["Country"] == "China"]
germany_df = df[df["Country"] == "Germany"]
mexico_df = df[df["Country"] == "Mexico"]
us_df = df[df["Country"] == "United States of America"]
zimbabwe_df = df[df["Country"] == "Zimbabwe"]

print(zimbabwe_df.head())
```

```python
plt.figure(figsize=(12,8))
plt.suptitle("Life Expectancy Changes (2000-2015)")
ax1 = plt.subplot(2,3,1)
ax1.plot(chile_df["Year"], chile_df["Life Expectancy"])
ax1.set_title("Chile")
ax2 = plt.subplot(2,3,2)
ax2.plot(china_df["Year"], china_df["Life Expectancy"])
ax2.set_title("China")
ax3 = plt.subplot(2,3,3)
ax3.plot(germany_df["Year"], germany_df["Life Expectancy"])
ax3.set_title("Germany")
ax4 = plt.subplot(2,3,4)
ax4.plot(mexico_df["Year"], mexico_df["Life Expectancy"])
ax4.set_title("Mexico")
ax5 = plt.subplot(2,3,5)
ax5.plot(us_df["Year"], us_df["Life Expectancy"])
ax5.set_title("United States")
ax6 = plt.subplot(2,3,6)
ax6.plot(zimbabwe_df["Year"], zimbabwe_df["Life Expectancy"])
ax6.set_title("Zimbabwe")

plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.suptitle("GDP Changes (2000-2015)")
ax1 = plt.subplot(2,3,1)
ax1.plot(chile_df["Year"], chile_df["GDP"])
ax1.set_title("Chile")
ax2 = plt.subplot(2,3,2)
ax2.plot(china_df["Year"], china_df["GDP"])
ax2.set_title("China")
ax3 = plt.subplot(2,3,3)
ax3.plot(germany_df["Year"], germany_df["GDP"])
ax3.set_title("Germany")
ax4 = plt.subplot(2,3,4)
ax4.plot(mexico_df["Year"], mexico_df["GDP"])
ax4.set_title("Mexico")
ax5 = plt.subplot(2,3,5)
ax5.plot(us_df["Year"], us_df["GDP"])
ax5.set_title("United States")
ax6 = plt.subplot(2,3,6)
ax6.plot(zimbabwe_df["Year"], zimbabwe_df["GDP"])
ax6.set_title("Zimbabwe")

plt.show()
```

```python
ax = sns.barplot(data=df, x="Country", y="Life Expectancy")
ax.set_title("Average Life Expectancy by Country")
ax.set_xticklabels(["Chile", "China", "Germany", "Mexico", "US", "Zimbabwe"])
plt.show()
```

```python
ax = sns.barplot(data=df, x="Country", y="GDP", hue="Year")
ax.set_title("Average GDP by Country")
ax.set_xticklabels(["Chile", "China", "Germany", "Mexico", "US", "Zimbabwe"])
plt.legend(ncol=3, fontsize="small")
ax.set_ylabel("GDP (Trillions)")
plt.show()
```

```python
ax = plt.subplot()
plt.scatter(df["Life Expectancy"], df["GDP"])
ax.set_ylabel("GDP (Trillions)")
ax.set_xlabel("Life Expectancy")
ax.set_title("Life Expectancy and GDP")
plt.show()

corr, pval = pearsonr(df["GDP"], df["Life Expectancy"])
print("Correlation between life expectancy and GDP is: " + str(corr))
print("The p-value is: " + str(pval))
```

```python
# Conclusions:
# Both life expectancy and GDP show positive trends for all countries involved over 2000-2015
# Chile, Germany, and Mexico show slight dips in GDP from 2012-2015, might be worth further research
# Germany has the highest average Life expectancy, Zimbabwe has the lowest. There is approximately a 20 year difference between the two.
# US has the highest GDP but China's GDP has risen the most drastically over the time period
# There does seem to be a linear relationship between life expectancy and GDP that is statistically significant.
#There is a wide range of life expectancy at Low GDP, but life expectancy in-line with high GDP is all grouped at the high end (73-80+).
```
